### Extracting data from a collection and creating a new object from the DataFrame class

```python
btc_data_dict = list(client.BTC.candlestick.find({}))

btc_data_df = pd.DataFrame.from_dict(btc_data_dict)
btc_data_df.head(2)
```

### Ploting!!

```python
candlestick = go.Figure(
    data=[go.Candlestick(x=btc_data_df['open_datetime'],
                         open=btc_data_df['open_price'],
                         high=btc_data_df['high_price'],
                         low=btc_data_df['low_price'],
                         close=btc_data_df['close_price'])]
)
candlestick.update_layout(paper_bgcolor='white', plot_bgcolor='white', margin={'b':10, 't':10, 'l':10, 'r':10}, width=1200, height=550)
candlestick.update_xaxes(showline=False, linecolor='black', gridcolor='lightgrey')
candlestick.update_yaxes(showline=False, linecolor='black', gridcolor='lightgrey')
candlestick.show()
```
