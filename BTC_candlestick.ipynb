# BTC Candlestick plot using Plotly

### Importing required dependencies

```python
import warnings
warnings.filterwarnings(action='ignore', category=DeprecationWarning)

from pymongo import MongoClient
from highcharts_stock.chart import Chart
import pandas as pd
import numpy as np
import config
```

### Establishing connection with mongoDB

```python
# Initializing mongoDB client
uri = f'mongodb+srv://josephnavarrete2:{config.MDB_PSWD}@btc-price-tracker.e3ejmb0.mongodb.net/?retryWrites=true&w=majority'
client = MongoClient(host=uri)
```

### Extracting data from a collection and creating a new object from the DataFrame class

```python
response = list(client.BTC.candlestick.find({}, {'_id': 0}))

btc_df = pd.DataFrame.from_dict(response)  # Parsing response from query to a pandas df just to transform it to a Numpy array
btc_arr = btc_df.to_numpy()
```

### Ploting!!

```python
as_dict = {
    'range_selector': {
        'selected': 1
    },
    'title': {
        'text': 'BTC/USDT Price'
    },
    'navigator': {
        'enabled': True
    },
    'series': [
        {
            'type': 'candlestick',
            'id': 'BTC',
            'name': 'BTC Price',
            'data': btc_arr
        }
    ],
    'plotOptions': {
        'candlestick': {
            'color': '#FF3254',
            'upColor': '#5BE93B'
        }
    }
}
```

```python
chart = Chart.from_options(as_dict)
chart.display()
```
